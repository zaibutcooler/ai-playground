```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return 2 * x + 2

# Define the interval [0, 2] and the number of subintervals
a = 0
b = 2
n = 4

# Calculate the width of each subinterval
delta_x = (b - a) / n

# Calculate the left endpoints of the subintervals
left_endpoints = np.linspace(a, b - delta_x, n)

# Evaluate the function at the left endpoints
f_values = f(left_endpoints)

# Compute the areas of the rectangles
areas = f_values * delta_x

# Plot the function
x_values = np.linspace(a, b, 100)
plt.plot(x_values, f(x_values), label='f(x) = 2x + 2')

# Plot the rectangles
for i in range(n):
    plt.bar(left_endpoints[i], f_values[i], width=delta_x, align='edge', alpha=0.5)

# Annotate the rectangles with their areas
for i in range(n):
    plt.text(left_endpoints[i] + delta_x / 2, f_values[i], f'{areas[i]:.2f}', ha='center', va='bottom')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Left-Hand Sum: Estimating Area under the Curve')
plt.legend()

# Show plot
plt.grid(True)
plt.show()

# Calculate the total estimated area
total_area = sum(areas)
print("Estimated area under the curve using left-hand sum:", total_area)
```

```python
import math

# Scenario 3: Find the 80th percentile for the sum of the 50 values of x

# Given parameters
mu_x = 45
sigma_x = 8
n = 50

# Given percentile
percentile_80 = 0.80

def standard_normal_cdf(z):
    return (1 + math.erf(z / math.sqrt(2))) / 2

def standard_normal_cdf_derivative(z):
    return math.exp(-z**2 / 2) / math.sqrt(2 * math.pi)

# Initialize variables for the Newton-Raphson method
tolerance = 1e-6  # Tolerance for convergence
guess = 0.0  # Initial guess
increment = 1.0  # Initial increment

# Implement the Newton-Raphson method to approximate the inverse normal CDF
while abs(standard_normal_cdf(guess) - percentile_80) > tolerance:
    guess -= increment * (standard_normal_cdf(guess) - percentile_80) / standard_normal_cdf_derivative(guess)

# Calculate the z-score corresponding to the 80th percentile
z_score_80th_percentile = guess

# Calculate the mean and standard deviation of the sum of the sample
mu_sum_x = mu_x * n
sigma_sum_x = sigma_x * math.sqrt(n)

# Calculate the sum of the sample values corresponding to the 80th percentile
sample_sum_80th_percentile = z_score_80th_percentile * sigma_sum_x + mu_sum_x

print("Scenario 3: 80th percentile for the sum of the 50 values of x:", sample_sum_80th_percentile)
```
