```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom

# Parameters
average_rate = 20  # Average rate of bug reports per week
num_trials = 1000  # Number of trials (e.g., software executions)
probability_crash = 0.05  # Probability of a crash in a single execution

# Calculate parameters for binomial distribution
n = num_trials
p = probability_crash

# Calculate parameters for Poisson distribution
lam = average_rate

# Generate x values (number of bug reports)
x = np.arange(0, 41)

# Calculate probabilities for binomial distribution
binomial_probs = binom.pmf(x, n, p)

# Calculate probabilities for Poisson distribution
poisson_probs = poisson.pmf(x, lam)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x, binomial_probs, label='Binomial Distribution (n={}, p={})'.format(n, p))
plt.plot(x, poisson_probs, 'r-', label='Poisson Distribution (={})'.format(lam))
plt.title('Comparison of Poisson and Binomial Distributions')
plt.xlabel('Number of Bug Reports')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

# Given parameters
average_rate = 20
k = 15

# Calculate Poisson probability
poisson_prob = (math.exp(-average_rate) * (average_rate ** k)) / math.factorial(k)

print(f"The probability of exactly 15 bug reports is approximately {poisson_prob * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Machine labels
machines = ['Machine 1', 'Machine 2']

# Mean values
mean_values = [50, 20]

# Standard deviation values
std_deviation = [49.5, 19.5]

# Plotting the bar graph
plt.bar(machines, mean_values, yerr=std_deviation, capsize=7)

# Adding labels and title
plt.xlabel('Machines')
plt.ylabel('Mean Transistors Produced')
plt.title('Mean Transistors Produced by Machines with Standard Deviation')

# Displaying the graph
plt.show()
```
