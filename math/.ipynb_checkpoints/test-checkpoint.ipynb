```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import sympy as smp
from sympy import *
```

```python
x = np.array([0,1,2,3,4,5])
y = np.array([1,2,6,10,26,12])
```

```python
# ratio or the shape of the graph
plt.figure(figsize=(8,3))

# add label
plt.xlabel('X label')
plt.ylabel('Y label')

plt.plot(x,y,'o--',color='purple')
```

```python
def f(x):
    return x*2

def g(x):
    return x**2

def assign(func):
    res = []
    for i in range(10):
        res.append(func(i))
          
    return res
```

```python
x = np.array(assign(f))
y = np.array(assign(g))
base_line = range(10)
```

```python
plt.figure(figsize=(8,4))
plt.plot(base_line,y,'--')
plt.plot(base_line,x,'o')
```

```python
class Animal():
    def __init__(self,name,eat,sleep):
        self.name = name
        self.eat = eat
        self.sleep = sleep

    def introduce(self):
        print(f"I am {self.name} and I eat {self.eat} and I sleep at {self.sleep}")
        
```

```python
dog = Animal(name="Dog",eat="dog food", sleep="floor")
dog.introduce()
```

```python
def arguments(*arg):
    for a in arg:
        print(f"{a}")
```
