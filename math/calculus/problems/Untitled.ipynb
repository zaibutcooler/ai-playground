```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def plot_triangle(base, height):
    # Define the function representing the height of the triangle as a function of x
    def h(x):
        return height * (1 - x / base)
    
    # Define the limits of integration
    a = 0
    b = base
    
    # Calculate the area of the triangle using integration
    area, _ = integrate.quad(h, a, b)
    
    # Plot the triangle
    x = np.linspace(a, b, 100)
    plt.plot(x, h(x), 'b', label=f'Triangle: base={base}, height={height}')
    plt.fill_between(x, h(x), color='blue', alpha=0.1)  # Fill the area under the triangle
    
    plt.title(f'Triangle: base={base}, height={height}, Area={area:.2f}')
    plt.xlabel('Base (x)')
    plt.ylabel('Height (y)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_triangle(4, 3)
```

```python
import sympy as sp

def calculate_triangle_area(base, height):
    # Define the variable and the function representing the height of the triangle as a function of x
    x = sp.Symbol('x')
    h = height * (1 - x / base)
    
    # Define the limits of integration
    a = 0
    b = base
    
    # Calculate the integral to find the area of the triangle
    area = sp.integrate(h, (x, a, b))
    
    return area

# Example usage
base = 4
height = 3
area = calculate_triangle_area(base, height)
print("Area of the triangle:", area)
```
