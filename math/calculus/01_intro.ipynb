It can find how things change overtime. Rate of change, **Slope** of a function.

#### Usecase of differentiation


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

t = sp.Symbol('t')

# Define the population function
population = 1000 * sp.exp(0.05 * t)

# Differentiate the population function to find the rate of population growth
population_growth_rate = sp.diff(population, t)

# Integrate the production rate function to find the accumulated total of items produced
production_rate = 10 + 5 * sp.sin(t/2)
total_production = sp.integrate(production_rate, (t, 0, 10))

# Create an array of time values from 0 to 10
time_values = np.linspace(0, 10, 100)

# Evaluate the population growth rate function at each time value
population_growth_rate_values = [population_growth_rate.evalf(subs={t: time}) for time in time_values]

# Evaluate the production rate function at each time value
production_rate_values = [production_rate.evalf(subs={t: time}) for time in time_values]

# Plotting the population growth rate function
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(time_values, population_growth_rate_values, label='Population Growth Rate')
plt.title('Population Growth Rate over Time')
plt.xlabel('Time')
plt.ylabel('Growth Rate')
plt.legend()

# Plotting the production rate function
plt.subplot(1, 2, 2)
plt.plot(time_values, production_rate_values, label='Production Rate')
plt.title('Production Rate over Time')
plt.xlabel('Time')
plt.ylabel('Production Rate')
plt.legend()

plt.tight_layout()
plt.show()

print("Total production during 0 to 10 units of time:", total_production)
```
