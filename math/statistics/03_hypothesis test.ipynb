## Hypothesis Testing

#### Why?
Hypothesis testing helps us make informed decisions, evaluate theories or claims, and advance scientific knowledge. It's used in various fields like medicine, economics, psychology, and more, to answer questions and solve problems based on data and evidence

#### What?
Hypothesis testing is like being a detective trying to solve a mystery about a population. You have a theory or a hunch (a hypothesis) about something in the population, and you want to see if the evidence from a sample supports or contradicts your theory

#### How?
It's not as simple as it sounds let's dive deep future billionare

#### Step 1: Formulate Hypotheses:

You start by stating your null and alternative hypotheses based on your research question or problem.

#### Step 2: Choose a Significance Level (α):

This is like setting your detective's standard for what counts as strong enough evidence. Common values are 0.05 or 0.01, meaning you're willing to accept a 5% or 1% chance of being wrong.

#### Step 3: Collect Data:

You collect data from a sample that represents the population you're interested in.

#### Step 4: Analyze Data:

You use statistical tests to analyze the data and calculate a test statistic, which is a number that summarizes the evidence in the data.

#### Step 5: Make a Decision:

You compare the test statistic to a critical value or calculate a p-value. If the test statistic falls in the critical region (or if the p-value is smaller than α), you reject the null hypothesis in favor of the alternative. Otherwise, you fail to reject the null hypothesis.

#### Step 6: Draw Conclusions:

Based on your decision, you draw conclusions about whether the evidence supports or contradicts your hypothesis. You also consider the practical significance of your findings.

### Simple Example

```python
# Step 1: Formulate Hypotheses
# Null Hypothesis (H0): The coin is fair (the probability of landing heads is 0.5)
# Alternative Hypothesis (Ha): The coin is biased (the probability of landing heads is not 0.5)

# Step 2: Choose a Significance Level (α)
alpha = 0.05

# Step 3: Collect Data (Fake Data for Demonstration)
import numpy as np

np.random.seed(42)  # for reproducibility
coin_flips = np.random.binomial(n=1, p=0.5, size=100)  # Simulate 100 coin flips (0 = tails, 1 = heads)

# Step 4: Analyze Data (Perform Binomial Test)
from scipy.stats import binom_test

num_heads = np.sum(coin_flips)
p_value = binom_test(num_heads, n=len(coin_flips), p=0.5)

# Step 5: Make a Decision
if p_value < alpha:
    print("Reject Null Hypothesis (H0)")
    print("The coin is biased.")
else:
    print("Fail to Reject Null Hypothesis (H0)")
    print("The coin is fair.")

# Step 6: Draw Conclusions
# Based on the decision, interpret the results and draw conclusions about whether the coin is fair or biased
```

### Types of testings

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, chi2_contingency, f_oneway, wilcoxon, mannwhitneyu

# Generate sample data
np.random.seed(42)

# t-test example
group1 = np.random.normal(loc=10, scale=2, size=100)
group2 = np.random.normal(loc=12, scale=2, size=100)
t_statistic, p_value = ttest_ind(group1, group2)
print("t-test p-value:", p_value)

# Chi-square test example
observed = pd.crosstab([1, 2, 3], [1, 2, 3])
chi2, p, _, _ = chi2_contingency(observed)
print("Chi-square test p-value:", p)

# ANOVA example
group1 = np.random.normal(loc=10, scale=2, size=100)
group2 = np.random.normal(loc=12, scale=2, size=100)
group3 = np.random.normal(loc=15, scale=2, size=100)
f_statistic, p_value = f_oneway(group1, group2, group3)
print("ANOVA p-value:", p_value)

# Paired t-test example
before = np.random.normal(loc=10, scale=2, size=100)
after = np.random.normal(loc=12, scale=2, size=100)
t_statistic, p_value = wilcoxon(before, after)
print("Paired t-test p-value:", p_value)

# Mann-Whitney U test example
group1 = np.random.normal(loc=10, scale=2, size=100)
group2 = np.random.normal(loc=12, scale=2, size=100)
u_statistic, p_value = mannwhitneyu(group1, group2)
print("Mann-Whitney U test p-value:", p_value)

# Plotting example
sns.histplot(data=[group1, group2], kde=True, stat="density")
plt.legend(["Group 1", "Group 2"])
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Distribution of Values")
plt.show()
```
