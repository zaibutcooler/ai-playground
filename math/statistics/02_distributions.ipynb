## Propability Distributions

```python
# Normal Distribution:
# A symmetric bell-shaped distribution representing many natural phenomena.
# Characterized by its mean and standard deviation.
# Follows the 68-95-99.7 rule (about 68% of data within one std deviation, 95% within two, and 99.7% within three).

# Binomial Distribution:
# Describes the number of successes in a fixed number of independent trials, each with the same probability of success.
# Used for binary outcomes (success/failure).
# Characterized by its number of trials (n) and probability of success (p).

# Poisson Distribution:
# Models the number of events occurring in a fixed interval of time or space.
# Used for rare events that occur independently at a constant rate.
# Characterized by its average rate (lambda).

# Exponential Distribution:
# Describes the time between events in a Poisson process, where events occur continuously and independently at a constant average rate.
# Used for modeling waiting times.
# Characterized by its rate parameter (lambda).

# Uniform Distribution:
# All outcomes are equally likely within a given range.
# Used to model situations where each outcome is equally probable.

# Gamma Distribution:
# A generalization of the exponential distribution used for modeling the waiting time until a specified number of events occur in a Poisson process.
# Characterized by its shape parameter (alpha) and scale parameter (beta).
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set(style="whitegrid")

# Generate fake data from each distribution
data_normal = np.random.normal(loc=0, scale=1, size=1000)
data_binomial = np.random.binomial(n=10, p=0.5, size=1000)
data_poisson = np.random.poisson(lam=5, size=1000)
data_exponential = np.random.exponential(scale=1/0.5, size=1000)
data_uniform = np.random.uniform(low=0, high=1, size=1000)
data_gamma = np.random.gamma(shape=2, scale=1, size=1000)

# Plotting
plt.figure(figsize=(12, 8))

# Normal Distribution
plt.subplot(3, 2, 1)
sns.histplot(data_normal, kde=True)
plt.title('Normal Distribution')

# Binomial Distribution
plt.subplot(3, 2, 2)
sns.histplot(data_binomial, kde=True)
plt.title('Binomial Distribution')

# Poisson Distribution
plt.subplot(3, 2, 3)
sns.histplot(data_poisson, kde=True)
plt.title('Poisson Distribution')

# Exponential Distribution
plt.subplot(3, 2, 4)
sns.histplot(data_exponential, kde=True)
plt.title('Exponential Distribution')

# Uniform Distribution
plt.subplot(3, 2, 5)
sns.histplot(data_uniform, kde=True)
plt.title('Uniform Distribution')

# Gamma Distribution
plt.subplot(3, 2, 6)
sns.histplot(data_gamma, kde=True)
plt.title('Gamma Distribution')

plt.tight_layout()
plt.show()
```
