### Continuity Distributions

It is used in uncountable(time).



#### Normal Distribution

Also known as gaussian distribution and a bell shape curve.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a seaborn histogram plot
sns.histplot(data, kde=True)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

#### Uniform Distribution



```python
# Generate random data from a uniform distribution
data = np.random.uniform(low=0, high=1, size=1000)

# Create a seaborn histogram plot
sns.histplot(data, kde=True)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate random data from an exponential distribution
data = np.random.exponential(scale=1, size=1000)

# Create a seaborn histogram plot
sns.histplot(data, kde=True)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate random data from a log-normal distribution
data = np.random.lognormal(mean=0, sigma=1, size=1000)

# Create a seaborn histogram plot
sns.histplot(data, kde=True)
plt.title('Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### Discrete Distributions

It is used in countable(card, dice).


#### Binomial Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data from a binomial distribution
data = np.random.binomial(n=10, p=0.5, size=1000)

# Create a seaborn histogram plot
sns.histplot(data, discrete=True, kde=False)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate random data from a Poisson distribution
data = np.random.poisson(lam=3, size=1000)

# Create a seaborn histogram plot
sns.histplot(data, discrete=True, kde=False)
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate random data from a geometric distribution
data = np.random.geometric(p=0.3, size=1000)

# Create a seaborn histogram plot
sns.histplot(data, discrete=True, kde=False)
plt.title('Geometric Distribution')
plt.xlabel('Number of Trials Until First Success')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate random data from a uniform discrete distribution
data = np.random.randint(low=1, high=7, size=1000)  # Rolling a fair six-sided die

# Create a seaborn histogram plot
sns.histplot(data, discrete=True, kde=False)
plt.title('Uniform Discrete Distribution')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()
```
