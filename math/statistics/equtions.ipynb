```python
import numpy as np

# Given data
x_values = [0, 1, 2, 3, 4, 5, 6]
probabilities = [0.12, 0.18, 0.30, 0.15, np.nan, 0.10, 0.05]

# Calculate the missing probability
missing_probability = 1 - np.sum([p for p in probabilities if not np.isnan(p)])
probabilities[probabilities.index(np.nan)] = missing_probability

# Calculate the expected value
expected_value = np.sum(np.array(x_values) * np.array(probabilities))

# Calculate the standard deviation
variance = np.sum(np.array(probabilities) * (np.array(x_values) - expected_value)**2)
standard_deviation = np.sqrt(variance)

# Print results
print("Completed Probabilities:", probabilities)
print("Expected Value:", expected_value)
print("Standard Deviation:", standard_deviation)
```
