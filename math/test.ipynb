```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 3**(x+2) + 5

# Generate x values
x_values = np.linspace(-5, 5, 400)

# Calculate corresponding y values
y_values = f(x_values)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$y = 3^{x+2} + 5$', color='black')
plt.title('Graph of $y = 3^{x+2} + 5$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.log(x+2) / np.log(3) + 5

# Generate x values
x_values = np.linspace(0.1, 10, 400)

# Calculate corresponding y values
y_values = f(x_values)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$y = \log_3(x+2) + 5$', color='black')
plt.title('Graph of $y = \log_3(x+2) + 5$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
X = [1, 2, 3, 4, 5, 6, 7]
Frequency = [2, 0, 8, 12, 12, 0, 1]

# Create bar plot
plt.bar(X, Frequency, color='skyblue')

# Add labels and title
plt.xlabel('Pairs of Sneakers Owned')
plt.ylabel('Frequency')
plt.title('Distribution of Pairs of Sneakers Owned')

# Add gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
import numpy as np

# Data
X = np.array([1, 2, 3, 4, 5, 6, 7])
Frequency = np.array([2, 0, 8, 12, 12, 0, 1])

# Mean
mean = np.sum(X * Frequency) / np.sum(Frequency)
print("Mean (x):", mean)

# Sample Standard Deviation
deviation = np.sqrt(np.sum(((X - mean) ** 2) * Frequency) / (np.sum(Frequency) - 1))
print("Sample Standard Deviation (s):", deviation)

# Quartiles
Q1 = np.percentile(X, 25, interpolation='lower')
median = np.percentile(X, 50, interpolation='lower')
Q3 = np.percentile(X, 75, interpolation='lower')
print("First Quartile (Q1):", Q1)
print("Median:", median)
print("Third Quartile (Q3):", Q3)

# Percentage of students owning at least five pairs
students_at_least_five = np.sum(Frequency[X >= 5])
total_students = np.sum(Frequency)
percentage_at_least_five = (students_at_least_five / total_students) * 100
print("Percentage of Students Owning At Least Five Pairs:", f"Approximately {percentage_at_least_five:.2f}%")

# 40th Percentile
percentile_40 = np.percentile(X, 40, interpolation='lower')
print("40th Percentile:", percentile_40)

# 90th Percentile
percentile_90 = np.percentile(X, 90, interpolation='lower')
print("90th Percentile:", percentile_90)
```
