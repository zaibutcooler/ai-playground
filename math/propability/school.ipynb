```python
import matplotlib.pyplot as plt

# Data
years = list(range(1961, 1974))
police = [260.35, 269.8, 272.04, 272.96, 272.51, 261.34, 268.89, 295.99, 319.87, 341.43, 356.59, 376.69, 390.19]
homicides = [8.6, 8.9, 8.52, 8.89, 13.07, 14.57, 21.36, 28.03, 31.49, 37.39, 46.26, 47.24, 52.33]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(years, police, marker='o', label='Police per 100,000 citizens')
plt.plot(years, homicides, marker='o', label='Homicides per 100,000 citizens')
plt.title('Number of Police vs. Homicides (1961-1973)')
plt.xlabel('Year')
plt.ylabel('Rate per 100,000 citizens')
plt.legend()
plt.grid(True)
plt.xticks(years, rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
years = list(range(1961, 1974))
police_per_100k = [260.35, 269.8, 272.04, 272.96, 272.51, 261.34, 268.89, 295.99, 319.87, 341.43, 356.59, 376.69, 390.19]
homicides_per_100k = [8.6, 8.9, 8.52, 8.89, 13.07, 14.57, 21.36, 28.03, 31.49, 37.39, 46.26, 47.24, 52.33]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(years, police_per_100k, marker='o', label='Police per 100k citizens')
plt.plot(years, homicides_per_100k, marker='o', label='Homicides per 100k citizens')
plt.xlabel('Year')
plt.ylabel('Rate per 100k citizens')
plt.title('Time Series of Police and Homicides (1961-1973)')
plt.legend()
plt.grid(True)

# Show plot
# plt.show()

# Calculate percentage increase relative to initial value
initial_police = police_per_100k[0]
initial_homicides = homicides_per_100k[0]
final_police = police_per_100k[-1]
final_homicides = homicides_per_100k[-1]

percentage_increase_police = ((final_police - initial_police) / initial_police) * 100
percentage_increase_homicides = ((final_homicides - initial_homicides) / initial_homicides) * 100

print("Percentage increase of police per 100k citizens:", percentage_increase_police)
print("Percentage increase of homicides per 100k citizens:", percentage_increase_homicides)

# Conclusion
if percentage_increase_police > percentage_increase_homicides:
    print("The variable 'police per 100k citizens' demonstrated a higher percentage increase relative to its initial value.")
else:
    print("The variable 'homicides per 100k citizens' demonstrated a higher percentage increase relative to its initial value.")
```
