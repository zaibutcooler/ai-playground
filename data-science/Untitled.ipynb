```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function y = sin(x) + cos(x)
def func(x):
    return np.sin(x) + np.cos(x)

# Define the range of x values
x_values = np.linspace(0, 2*np.pi, 1000)

# Plot the original function y = sin(x) + cos(x)
plt.plot(x_values, func(x_values), label='y = sin(x) + cos(x)')

# Plot the function y = A*sin(x) + B*cos(x) for A = 2 and B = 1
A1, B1 = 2, 1
plt.plot(x_values, A1*np.sin(x_values) + B1*np.cos(x_values), label=f'y = {A1}*sin(x) + {B1}*cos(x)')

# Plot the function y = A*sin(x) + B*cos(x) for A = 1 and B = 2
A2, B2 = 1, 2
plt.plot(x_values, A2*np.sin(x_values) + B2*np.cos(x_values), label=f'y = {A2}*sin(x) + {B2}*cos(x)')

# Plot maximum points
plt.scatter(np.pi/4, np.sqrt(2), color='red', label='Maximum Point (Original)')
plt.scatter(np.pi/4, 2*np.sqrt(2), color='blue', label='Maximum Point (A=2, B=1)')
plt.scatter(np.pi/4, 2*np.sqrt(2), color='green', label='Maximum Point (A=1, B=2)')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of Trigonometric Functions')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Define the sizes of sets A and B and their intersection
A_size = 750
B_size = 900
A_intersect_B = 530

# Create a Venn diagram
venn = venn2(subsets=(A_size - A_intersect_B, B_size - A_intersect_B, A_intersect_B),
             set_labels=('People in IT Sector', 'People with Jobs'))

# Customize the diagram
venn.get_label_by_id('10').set_text(f'Only in IT Sector: {A_size - A_intersect_B}')
venn.get_label_by_id('01').set_text(f'Only with Jobs: {B_size - A_intersect_B}')
venn.get_label_by_id('11').set_text(f'In both: {A_intersect_B}')

# Add title
plt.title("Venn Diagram - People in IT Sector vs People with Jobs")

# Show the plot
plt.show()
```
