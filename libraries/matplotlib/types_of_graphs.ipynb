```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.arange(10)
y = np.random.randint(1, 10, size=10)
categories = ['A', 'B', 'C', 'D', 'E']

# Line Graph
plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title('Line Graph Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# Bar Graph
plt.figure(figsize=(8, 5))
plt.bar(categories, y)
plt.title('Bar Graph Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Histogram
plt.figure(figsize=(8, 5))
plt.hist(y, bins=5)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Pie Chart
plt.figure(figsize=(8, 5))
plt.pie(y, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()

# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# Box Plot
plt.figure(figsize=(8, 5))
plt.boxplot(y)
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()

# Heatmap
plt.figure(figsize=(8, 5))
plt.imshow([y], cmap='hot', aspect='auto')
plt.colorbar()
plt.title('Heatmap Example')
plt.show()

# Area Graph
plt.figure(figsize=(8, 5))
plt.fill_between(x, y)
plt.title('Area Graph Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Bubble Chart
bubble_sizes = np.random.randint(100, 1000, size=len(x))
plt.figure(figsize=(8, 5))
plt.scatter(x, y, s=bubble_sizes, alpha=0.5)
plt.title('Bubble Chart Example')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Pareto Chart
sorted_categories = [x for _, x in sorted(zip(y, categories), reverse=True)]
sorted_y = sorted(y, reverse=True)
plt.figure(figsize=(8, 5))
plt.bar(sorted_categories, sorted_y, color='b')
cumulative_y = np.cumsum(sorted_y)
plt.plot(sorted_categories, cumulative_y, color='r', marker='o')
plt.title('Pareto Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend(['Cumulative', 'Category'])
plt.show()
```
